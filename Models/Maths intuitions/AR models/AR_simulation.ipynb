# Simulation of autoregressive processes 

```python
"""
IMPORTS
"""
import numpy as np
import numpy.random as npr
import os
import random
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

npr.seed(65)
```

### Parameters

```python
def unroll(y_prev, A, b, sigma):

    mu = y_prev * A + b
    y_new = np.random.normal(mu, sigma, 1)

    return y_new


def autoregressive (y_zero, A, b, sigma, len):

    y = np.array([])
    i = 0
    y_prev = y_zero
    while i < len:
        y_new = unroll(y_prev, A, b, sigma)
        y = np.append(y, np.array([y_new]))
        y_prev = y_new
        i += 1

    return y
```

```python
A = .3
b = 2
sigma = .01
y_zero = 0
len = 100
y = autoregressive (y_zero, A, b, sigma, len)
plt.plot(y)
```
